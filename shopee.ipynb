```python
class Shopee:     
    def __init__(self, page, Search = "11043145"):
        self.ENDPOINT = f"https://shopee.co.id/api/v4/search/search_items?by=relevancy&limit=60&match_id={Search}&newest=1860&order=desc&page_type=search&scenario=PAGE_GLOBAL_SEARCH&version=2"
        self.SHOP = "https://shopee.co.id/api/v4/product/get_shop_info?shopid="
    def getShopee(self) :
        sorting          = []
        request_products = json.loads(req.get(self.ENDPOINT).content)
        for indexProducts in range(len(request_products["items"])) :
            requestShop  = json.loads(req.get("https://shopee.co.id/api/v4/product/get_shop_info?shopid="+str(request_products["items"][indexProducts]["item_basic"]["shopid"])).content)
            def urlProduk():
                endpoint = "https://shopee.co.id/-"
                Name     = request_products["items"][indexProducts]["item_basic"]["name"]
                result   = ""
                shopid   = str(request_products["items"][indexProducts]["item_basic"]["shopid"])
                itemid   = str(request_products["items"][indexProducts]["item_basic"]["itemid"])
                for indexResult in range(len(Name)):
                    if format(ord(Name[indexResult]), "x") == "20":
                        result+="-"
                    else:
                        result+=Name[indexResult]
                return f"{endpoint}{result}-i.{shopid}.{itemid}"
            Profile = {
                "uname_toko"  : requestShop["data"]["account"]["username"],
                "nama_toko"   : requestShop["data"]["name"],
                "lokasi_toko" : requestShop["data"]["shop_location"],
                "jumlah_prod" : requestShop["data"]["item_count"],
                "jumlah_rating" : requestShop["data"]["rating_star"],
                "performa_chat" : requestShop["data"]["response_rate"],
                "waktu_chat"  : requestShop["data"]["response_time"],
                "follower"    : requestShop["data"]["follower_count"],
                "rating_bad"  : requestShop["data"]["rating_bad"],
                "rating_good" : requestShop["data"]["rating_good"],
                "rating_normal": requestShop["data"]["rating_normal"],
                "status_verified": requestShop["data"]["is_shopee_verified"],
                "status_plus": requestShop["data"]["is_preferred_plus_seller"],
                "status_official": requestShop["data"]["is_official_shop"],
                "join_time"   : requestShop["data"]["ctime"],
                "url_toko"    : "https://shopee.co.id/"+requestShop["data"]["account"]["username"]
            }
            sorting.append(
                Profile
            )

        return sorting
```
